# 06. 集合

"paraparaparadise"と"paragraph"に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，'se'というbi-gramがXおよびYに含まれるかどうかを調べよ．

```
# 問題06: 集合

# n-gramを作成する関数（問題05で作成した関数を再利用）
def n_gram(sequence, n):
    """シーケンスからn-gramを作成する関数
    
    Args:
        sequence: 文字列またはリスト
        n: n-gramのn
        
    Returns:
        n-gramのリスト
    """
    return [sequence[i:i+n] for i in range(len(sequence) - n + 1)]

# 与えられた文字列
str1 = "paraparaparadise"
str2 = "paragraph"

# 文字bi-gramの集合を作成
X = set(n_gram(str1, 2))
Y = set(n_gram(str2, 2))

# 和集合、積集合、差集合を計算
union_set = X | Y  # または X.union(Y)
intersection_set = X & Y  # または X.intersection(Y)
difference_set_X_Y = X - Y  # または X.difference(Y)
difference_set_Y_X = Y - X  # または Y.difference(X)

# 'se'がXおよびYに含まれるかどうかを調べる
se_in_X = "se" in X
se_in_Y = "se" in Y

# 結果を表示
print(f"文字列1: {str1}")
print(f"文字列2: {str2}")
print(f"X: {X}")
print(f"Y: {Y}")
print(f"和集合 (X | Y): {union_set}")
print(f"積集合 (X & Y): {intersection_set}")
print(f"差集合 (X - Y): {difference_set_X_Y}")
print(f"差集合 (Y - X): {difference_set_Y_X}")
print(f"'se' in X: {se_in_X}")
print(f"'se' in Y: {se_in_Y}")
```

## 解説

この問題では、2つの文字列から文字bi-gramの集合を作成し、集合演算を行います。

1. まず、問題05で作成したn-gram関数を使用して、各文字列の文字bi-gramを生成します。
2. 次に、これらのbi-gramをセット（集合）に変換します。
3. 集合演算（和集合、積集合、差集合）を行います。
4. 最後に、特定のbi-gram（'se'）が各集合に含まれるかどうかを確認します。

Pythonでは、集合演算を行うための演算子が用意されています：
- 和集合（Union）: `|` または `union()`メソッド
- 積集合（Intersection）: `&` または `intersection()`メソッド
- 差集合（Difference）: `-` または `difference()`メソッド

集合は重複を許さないため、同じbi-gramが複数回出現しても、集合内では1つとしてカウントされます。
