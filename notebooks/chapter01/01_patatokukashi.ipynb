# 01. 「パタトクカシーー」

「パタトクカシーー」という文字列の1,3,5,7文字目を取り出して連結した文字列を得よ．

````
# 問題01: 「パタトクカシーー」

# 与えられた文字列
s = "パタトクカシーー"

# 1,3,5,7文字目を取り出して連結
# 注意: Pythonのインデックスは0から始まるため、1,3,5,7文字目は実際にはインデックス0,2,4,6
result = s[0] + s[2] + s[4] + s[6]

# 結果を表示
print(f"元の文字列: {s}")
print(f"抽出結果: {result}")
````

## 解説

Pythonでは、文字列の各文字にインデックスを使ってアクセスできます。インデックスは0から始まるため、1文字目はインデックス0、3文字目はインデックス2、というように対応します。

- 1文字目（インデックス0）: パ
- 3文字目（インデックス2）: ト
- 5文字目（インデックス4）: カ
- 7文字目（インデックス6）: ー

これらを連結すると「パトカー」という文字列が得られます。

別の解法として、スライシングを使用することもできます：
```python
result = s[::2]  # 開始位置から終了位置まで、2文字ごとに取得
```
この方法でも同じ結果が得られます。
