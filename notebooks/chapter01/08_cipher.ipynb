# 08. 暗号文

与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．

- 英小文字ならば(219 - 文字コード)の文字に置換
- その他の文字はそのまま出力

この関数を用い，英語のメッセージを暗号化・復号化せよ．

```
# 問題08: 暗号文

# 暗号化・復号化関数
def cipher(text):
    """文字列を暗号化・復号化する関数
    
    英小文字ならば(219 - 文字コード)の文字に置換し、
    その他の文字はそのまま出力する
    
    Args:
        text: 入力文字列
        
    Returns:
        暗号化・復号化された文字列
    """
    result = ""
    for char in text:
        if char.islower():  # 英小文字かどうかを判定
            # 文字コードを取得し、変換式を適用
            code = ord(char)
            converted_code = 219 - code
            result += chr(converted_code)
        else:
            # その他の文字はそのまま
            result += char
    return result

# テスト用の文字列
original_text = "Hello, World! This is a test message."

# 暗号化
encrypted_text = cipher(original_text)

# 復号化（同じ関数を適用）
decrypted_text = cipher(encrypted_text)

# 結果を表示
print(f"元の文字列: {original_text}")
print(f"暗号化後: {encrypted_text}")
print(f"復号化後: {decrypted_text}")
```

## 解説

この問題では、特定のルールに従って文字列を変換する暗号化・復号化関数を実装します。

変換ルールは以下の通りです：
- 英小文字（a-z）の場合、(219 - 文字コード)の文字に置換します。
- その他の文字（大文字、数字、記号、空白など）はそのまま出力します。

Pythonでは、`ord()`関数を使用して文字のUnicode文字コードを取得し、`chr()`関数を使用して文字コードから文字を取得できます。また、`islower()`メソッドを使用して、文字が英小文字かどうかを判定できます。

例えば、'a'の文字コードは97なので、変換後は219 - 97 = 122となり、これは'z'の文字コードです。同様に、'z'の文字コードは122なので、変換後は219 - 122 = 97となり、これは'a'の文字コードです。

興味深いことに、この暗号化方式は対称的であり、同じ関数を2回適用すると元の文字列に戻ります。これは、(219 - (219 - x)) = xとなるためです。
