# 09. Typoglycemia

スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．ただし，長さが4以下の単語は並び替えないこととする．適当な英語の文（例えば"I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."）を与え，その実行結果を確認せよ．

```
# 問題09: Typoglycemia

import random

# Typoglycemiaを実装する関数
def typoglycemia(text):
    """単語の先頭と末尾の文字は残し、それ以外の文字の順序をランダムに並び替える関数
    
    ただし、長さが4以下の単語は並び替えない
    
    Args:
        text: 入力文字列
        
    Returns:
        変換後の文字列
    """
    words = text.split()
    result = []
    
    for word in words:
        # 句読点を分離
        if word[-1] in ['.', ',', ':', ';', '!', '?']:
            punct = word[-1]
            word = word[:-1]
        else:
            punct = ''
        
        # 長さが4以下の単語は並び替えない
        if len(word) <= 4:
            result.append(word + punct)
        else:
            # 先頭と末尾の文字を保持し、それ以外の文字をランダムに並び替え
            middle_chars = list(word[1:-1])
            random.shuffle(middle_chars)
            shuffled_word = word[0] + ''.join(middle_chars) + word[-1] + punct
            result.append(shuffled_word)
    
    return ' '.join(result)

# テスト用の文
text = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."

# シード値を固定して再現性を確保
random.seed(42)

# Typoglycemiaを適用
shuffled_text = typoglycemia(text)

# 結果を表示
print(f"元の文: {text}")
print(f"変換後: {shuffled_text}")
```

## 解説

この問題では、「Typoglycemia」と呼ばれる現象を模倣するプログラムを実装します。Typoglycemiaとは、単語の先頭と末尾の文字が正しい位置にあれば、中間の文字の順序が入れ替わっていても人間は読むことができるという現象です。

実装した関数では、以下の処理を行っています：

1. 入力文字列をスペースで区切って単語のリストに分割します。
2. 各単語に対して：
   - 句読点（.,:;!?）が末尾にある場合、一時的に分離します。
   - 単語の長さが4以下の場合、そのまま出力します。
   - 単語の長さが5以上の場合、先頭と末尾の文字を保持し、中間の文字をランダムに並び替えます。
3. 処理した単語を再びスペースで連結して返します。

ランダムな結果を再現可能にするために、`random.seed()`を使用して乱数生成器のシード値を固定しています。

このプログラムを実行すると、長い単語の中間部分がランダムに並び替えられますが、先頭と末尾の文字は変わらないため、文章全体の意味は理解できる状態を保ちます。
